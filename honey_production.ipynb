```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
file_to_load = "Resources/Honey_State_Production.csv"
file2_to_load = "Resources/National_Honey_Sales.csv"
file3_to_load = "Resources/Honey_US_Production.csv"
file4_to_load = "Resources/Honey_Survey_Prices.csv"
state_production_data = pd.read_csv(file_to_load)
census_price_data = pd.read_csv(file2_to_load)
national_production_data = pd.read_csv(file3_to_load)
survey_price_data = pd.read_csv(file4_to_load)
```

```python
state_honey_df = pd.DataFrame(state_production_data)
national_honey_df = pd.DataFrame(national_production_data)
survey_prices = pd.DataFrame(survey_price_data)
census_prices = pd.DataFrame(census_price_data)
```

```python
state_honey_df["Value"] = pd.to_numeric(state_honey_df["Value"].str.replace(',',''), errors='coerce')
```

```python
Texas_honey = state_honey_df.loc[(state_honey_df["State"] == "TEXAS")]
Texas_production = Texas_honey[["Year" , "Value"]]
Texas_production
```

```python
production_plot = Texas_production.plot.line(x='Year', y='Value', xticks= Texas_production['Year'])
production_plot.set_title("Texas Honey Production 1997 - 2017 (USDA Census)")
production_plot.set_xlabel("Year")
production_plot.set_ylabel("Production (per 100,000,000 lbs)")
```

```python
national_honey_df["Value"] = pd.to_numeric(national_honey_df["Value"].str.replace(',',''), errors='coerce')
```

```python
national_production = national_honey_df[["Year" , "Value"]]
national_production
```

```python
US_production_plot = national_production.plot.line(x='Year', y='Value', xticks= national_production['Year'])
US_production_plot.set_title("U.S. Honey Production 1997 - 2017 (USDA Census)")
US_production_plot.set_xlabel("Year")
US_production_plot.set_ylabel("Production (per 100,000,000 lbs)")
```

```python
National_prices = survey_prices.loc[(survey_prices["State"] == "US TOTAL")]
National_prices
```

```python
Prices = National_prices[["Year","State","Value"]]
```

```python
Real_Prices = round(Prices["Value"]/100,2)
Prices["Value"] = Real_Prices

```

```python
# price_plot = Prices.plot.line(x='Year', y='Value', xticks= Prices['Year'])
price_plot = Prices.plot.line(x='Year', y='Value')
price_plot.set_title("U.S. Honey Prices 1987 - 2018 (USDA Survey)")
price_plot.set_xlabel("Year")
price_plot.set_ylabel("Price per $1 USD")
```

```python
Texas_prices = survey_prices.loc[(survey_prices["State"] == "TEXAS")]
Texas_prices
```

```python
T_Prices = Texas_prices[["Year", "State", "Value"]]
```
